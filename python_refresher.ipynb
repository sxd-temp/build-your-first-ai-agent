## 1 · Variables — a name for a value
A variable is a labeled box: a **name** on the left, a **value** on the right. Everything else builds on this — it's how we hold the things we work with. In the labs you'll mostly just *change the value*.

```python
city = "Lisbon"        # a piece of text (a "string")
temperature = 0.7      # a number

print(city)
print(temperature)
```

## 2 · Strings — text inside quotes
Text goes inside quotes. Single `'...'` or double `"..."` both work. For text that spans several lines, use **triple quotes** `"""..."""`.

⚠️ Every opening quote needs a closing one — a missing quote is the **#1 typo**. The `role`, `goal`, and `backstory` you edit later are all just strings like these.

```python
short_note = "keep it simple"

long_note = """This note
runs across
three lines."""

print(short_note)
print(long_note)
```

## 3 · Comments — notes Python ignores
Anything after a `#` is a note for humans; Python skips it. Comments explain code — they never "run".

```python
price = 5        # the # starts a comment; Python ignores it
# this whole line is a comment, too

print(price)
```

## 4 · Lists — several items, in order
A list holds several items inside square brackets `[ ]`, separated by commas. It can hold strings, numbers — or nothing at all (an **empty list** `[]`). You'll pass lists around in the labs: sometimes a full one, sometimes an empty `[]`.

```python
toppings = ["mushroom", "olives", "basil"]
no_toppings = []

print(toppings)
print(len(toppings))   # len() counts the items -> 3
print(no_toppings)
```

## 5 · Functions — *calling* one
A function is a reusable action with a name. You **call** it by writing its name followed by parentheses `()`. You've already used two: `print(...)` and `len(...)`.

The values inside the parentheses are **arguments** — the inputs you hand the function.

```python
count = len([10, 20, 30])   # len is a function; it returns (gives back) a value
print(count)
```

## 6 · Keyword arguments — naming the inputs ⭐
When a function takes several inputs, you can **name** each one: `name=value`. That's a **keyword argument** — it makes the call readable and order-proof.

**This is the exact shape you'll edit all day in the labs** — so give it a good look.

```python
def order_coffee(size, shots, oat_milk):
    return f"{size} coffee · {shots} shots · oat milk: {oat_milk}"

# called with keyword arguments — each input is named:
my_order = order_coffee(size="large", shots=2, oat_milk=True)
print(my_order)
```

## 7 · Reading a function definition
You won't *write* functions today — but you'll *read* them, so here's the shape:
`def` names it · the **indented** lines are its body · `return` is what it hands back.

Indentation matters in Python — the body is indented under the `def`. Keep it consistent and you're fine.

```python
def greet(name):
    message = "Hi, " + name + "!"
    return message

print(greet("Sam"))
```

## 8 · Putting it together — the shape you'll see in the labs
Here's everything at once, in a friendly example: a function that builds a **game character** from named inputs, including a **list**.

Read it like a sentence — a **name** (string), a **role** (string), and **skills** (a list). The lab code uses this *exact skeleton* — a thing configured with keyword arguments and a list — just in a different story.

```python
def create_character(name, role, skills):
    return {"name": name, "role": role, "skills": skills}

hero = create_character(
    name="Mara",
    role="A brave healer who never leaves a teammate behind.",   # a string
    skills=["heal", "shield", "revive"],                         # a list
)

print(hero)
```

## 9 · The 3 rules that prevent almost every error
1. **Balanced quotes** — every `"` needs a closing `"`.
2. **Commas between inputs** — `f(a=1, b=2)`, not `f(a=1 b=2)`.
3. **Consistent indentation** — lines that belong together start at the same spot.

If something breaks, it's almost always one of these three. And remember: select the red error and ask **Copilot** *"what's wrong with this?"*

```python
# correct: balanced quotes, commas between inputs, consistent indentation
print(order_coffee(size="small", shots=1, oat_milk=False))
```
